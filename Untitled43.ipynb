```python
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
import openpyxl
from openpyxl import Workbook
import os
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

# File to store the data
FILE_NAME = "Student_Details.xlsx"

# Email settings
SENDER_EMAIL = "youremail@gmail.com"  # Replace with your email
SENDER_PASSWORD = "yourpassword"  # Replace with your email password or app password
RECIPIENT_EMAIL = "recipient@example.com"  # Replace with recipient's email

# Function to send email with student details
def send_email(name, roll_number, age, branch):
    try:
        msg = MIMEMultipart()
        msg['From'] = SENDER_EMAIL
        msg['To'] = RECIPIENT_EMAIL
        msg['Subject'] = 'Student Details Submission Receipt'

        body = f"""
        Dear User,

        Your contact details have been updated successfully.

        Name: {name}
        Roll Number: {roll_number}
        Age: {age}
        Branch: {branch}

        Thank you for submitting the form!

        Regards,
        Your School/Institute
        """

        msg.attach(MIMEText(body, 'plain'))

        # Establishing the SMTP server connection and sending the email
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(SENDER_EMAIL, SENDER_PASSWORD)
        text = msg.as_string()
        server.sendmail(SENDER_EMAIL, RECIPIENT_EMAIL, text)
        server.quit()

        messagebox.showinfo("Email Sent", "Receipt has been sent to your email.")
    except Exception as e:
        messagebox.showerror("Error", f"Failed to send email. Error: {str(e)}")

# Function to generate a PDF receipt
def generate_pdf(name, roll_number, age, branch):
    pdf_filename = f"Student_Receipt_{roll_number}.pdf"
    c = canvas.Canvas(pdf_filename, pagesize=letter)

    # Title
    c.setFont("Helvetica-Bold", 14)
    c.drawString(180, 750, "Student Details Submission Receipt")

    # Student details
    c.setFont("Helvetica", 12)
    c.drawString(100, 700, f"Name: {name}")
    c.drawString(100, 680, f"Roll Number: {roll_number}")
    c.drawString(100, 660, f"Age: {age}")
    c.drawString(100, 640, f"Branch: {branch}")

    # Saving the PDF
    c.save()

    messagebox.showinfo("PDF Generated", f"PDF receipt has been saved as {pdf_filename}.")
    return pdf_filename

# Function to save student details to Excel
def save_to_excel(name, roll_number, age, branch):
    if not os.path.exists(FILE_NAME):
        # Create a new workbook and add a header row
        workbook = Workbook()
        sheet = workbook.active
        sheet.title = "Details"
        sheet.append(["Name", "Roll Number", "Age", "Branch"])
        workbook.save(FILE_NAME)

    # Open the workbook
    workbook = openpyxl.load_workbook(FILE_NAME)
    sheet = workbook.active

    # Check for duplicate Roll Number
    for row in sheet.iter_rows(min_row=2, values_only=True):
        if row[1] == roll_number:
            return False

    # Append the data
    sheet.append([name, roll_number, age, branch])  # Add data row
    workbook.save(FILE_NAME)
    return True

# Function to collect and display student details
def submit_details():
    name = name_entry.get()
    roll_number = roll_entry.get()
    age = age_entry.get()
    branch = branch_combobox.get()

    if not (name and roll_number and age and branch) or branch == "Select Branch":
        messagebox.showwarning("Input Error", "All fields are required!")
        return

    # Save details to Excel and check for duplicates
    if save_to_excel(name, roll_number, age, branch):
        messagebox.showinfo("Success", "Your contact has been updated!")

        # Send email and generate PDF
        send_email(name, roll_number, age, branch)
        generate_pdf(name, roll_number, age, branch)

        clear_fields()
    else:
        messagebox.showwarning("Duplicate Entry", "This record already exists!")

# Function to clear input fields
def clear_fields():
    name_entry.delete(0, tk.END)
    roll_entry.delete(0, tk.END)
    age_entry.delete(0, tk.END)
    branch_combobox.set("Select Branch")

# Create main window
root = tk.Tk()
root.title("Student Details Form")
root.geometry("400x450")
root.configure(bg="#f0f0f0")  # Set background color

# Create a frame for the form elements
form_frame = tk.Frame(root, bg="#f0f0f0")
form_frame.pack(pady=20)

# Add header label
header_label = tk.Label(form_frame, text="Student Details Form", font=("Helvetica", 16, "bold"), bg="#f0f0f0")
header_label.grid(row=0, column=0, columnspan=2, pady=10)

# Create labels and entry fields with styling
tk.Label(form_frame, text="Name:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=1, column=0, padx=10, pady=5, sticky="w")
name_entry = tk.Entry(form_frame, font=("Helvetica", 12), width=30, bd=2)
name_entry.grid(row=1, column=1, pady=5)

tk.Label(form_frame, text="Roll Number:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=2, column=0, padx=10, pady=5, sticky="w")
roll_entry = tk.Entry(form_frame, font=("Helvetica", 12), width=30, bd=2)
roll_entry.grid(row=2, column=1, pady=5)

tk.Label(form_frame, text="Age:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=3, column=0, padx=10, pady=5, sticky="w")
age_entry = tk.Entry(form_frame, font=("Helvetica", 12), width=30, bd=2)
age_entry.grid(row=3, column=1, pady=5)

tk.Label(form_frame, text="Branch:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=4, column=0, padx=10, pady=5, sticky="w")
branch_combobox = ttk.Combobox(form_frame, font=("Helvetica", 12), width=28, state="readonly")
branch_combobox['values'] = ("EXTC", "IT", "CS", "Mechanical", "Civil")
branch_combobox.set("Select Branch")
branch_combobox.grid(row=4, column=1, pady=5)

# Buttons with styling
button_frame = tk.Frame(root, bg="#f0f0f0")
button_frame.pack(pady=10)

submit_button = tk.Button(button_frame, text="Submit", font=("Helvetica", 12, "bold"), bg="#4CAF50", fg="white", bd=2, relief="raised", width=12, command=submit_details)
submit_button.grid(row=0, column=0, padx=10)

clear_button = tk.Button(button_frame, text="Clear", font=("Helvetica", 12, "bold"), bg="#f44336", fg="white", bd=2, relief="raised", width=12, command=clear_fields)
clear_button.grid(row=0, column=1, padx=10)

# Run the GUI event loop
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
import openpyxl
from openpyxl import Workbook
import os
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

# File to store the data
FILE_NAME = "Student_Details.xlsx"

# Email settings
SENDER_EMAIL = "youremail@gmail.com"  # Replace with your email
SENDER_PASSWORD = "yourpassword"  # Replace with your email password or app password
RECIPIENT_EMAIL = "recipient@example.com"  # Replace with recipient's email

# Global variable to store the generated PDF file path
pdf_filename = ""

# Function to send email with student details
def send_email(name, roll_number, age, branch):
    try:
        msg = MIMEMultipart()
        msg['From'] = SENDER_EMAIL
        msg['To'] = RECIPIENT_EMAIL
        msg['Subject'] = 'Student Details Submission Receipt'

        body = f"""
        Dear User,

        Your contact details have been updated successfully.

        Name: {name}
        Roll Number: {roll_number}
        Age: {age}
        Branch: {branch}

        Thank you for submitting the form!

        Regards,
        Your School/Institute
        """

        msg.attach(MIMEText(body, 'plain'))

        # Establishing the SMTP server connection and sending the email
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(SENDER_EMAIL, SENDER_PASSWORD)
        text = msg.as_string()
        server.sendmail(SENDER_EMAIL, RECIPIENT_EMAIL, text)
        server.quit()

        messagebox.showinfo("Email Sent", "Receipt has been sent to your email.")
    except Exception as e:
        messagebox.showerror("Error", f"Failed to send email. Error: {str(e)}")

# Function to generate a PDF receipt
def generate_pdf(name, roll_number, age, branch):
    global pdf_filename
    pdf_filename = f"Student_Receipt_{roll_number}.pdf"
    c = canvas.Canvas(pdf_filename, pagesize=letter)

    # Title
    c.setFont("Helvetica-Bold", 14)
    c.drawString(180, 750, "Student Details Submission Receipt")

    # Student details
    c.setFont("Helvetica", 12)
    c.drawString(100, 700, f"Name: {name}")
    c.drawString(100, 680, f"Roll Number: {roll_number}")
    c.drawString(100, 660, f"Age: {age}")
    c.drawString(100, 640, f"Branch: {branch}")

    # Saving the PDF
    c.save()

    messagebox.showinfo("PDF Generated", f"PDF receipt has been saved as {pdf_filename}.")
    return pdf_filename

# Function to save student details to Excel
def save_to_excel(name, roll_number, age, branch):
    if not os.path.exists(FILE_NAME):
        # Create a new workbook and add a header row
        workbook = Workbook()
        sheet = workbook.active
        sheet.title = "Details"
        sheet.append(["Name", "Roll Number", "Age", "Branch"])
        workbook.save(FILE_NAME)

    # Open the workbook
    workbook = openpyxl.load_workbook(FILE_NAME)
    sheet = workbook.active

    # Check for duplicate Roll Number
    for row in sheet.iter_rows(min_row=2, values_only=True):
        if row[1] == roll_number:
            return False

    # Append the data
    sheet.append([name, roll_number, age, branch])  # Add data row
    workbook.save(FILE_NAME)
    return True

# Function to collect and display student details
def submit_details():
    name = name_entry.get()
    roll_number = roll_entry.get()
    age = age_entry.get()
    branch = branch_combobox.get()

    if not (name and roll_number and age and branch) or branch == "Select Branch":
        messagebox.showwarning("Input Error", "All fields are required!")
        return

    # Save details to Excel and check for duplicates
    if save_to_excel(name, roll_number, age, branch):
        messagebox.showinfo("Success", "Your contact has been updated!")

        # Send email and generate PDF
        send_email(name, roll_number, age, branch)
        generate_pdf(name, roll_number, age, branch)

        clear_fields()
        download_button.grid(row=5, column=1, pady=10)  # Show the download button
    else:
        messagebox.showwarning("Duplicate Entry", "This record already exists!")

# Function to clear input fields
def clear_fields():
    name_entry.delete(0, tk.END)
    roll_entry.delete(0, tk.END)
    age_entry.delete(0, tk.END)
    branch_combobox.set("Select Branch")

# Function to download the PDF receipt
def download_receipt():
    if pdf_filename:
        messagebox.showinfo("Download Receipt", f"Download your receipt from: {os.path.abspath(pdf_filename)}")

# Create main window
root = tk.Tk()
root.title("Student Details Form")
root.geometry("400x500")
root.configure(bg="#f0f0f0")  # Set background color

# Create a frame for the form elements
form_frame = tk.Frame(root, bg="#f0f0f0")
form_frame.pack(pady=20)

# Add header label
header_label = tk.Label(form_frame, text="Student Details Form", font=("Helvetica", 16, "bold"), bg="#f0f0f0")
header_label.grid(row=0, column=0, columnspan=2, pady=10)

# Create labels and entry fields with styling
tk.Label(form_frame, text="Name:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=1, column=0, padx=10, pady=5, sticky="w")
name_entry = tk.Entry(form_frame, font=("Helvetica", 12), width=30, bd=2)
name_entry.grid(row=1, column=1, pady=5)

tk.Label(form_frame, text="Roll Number:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=2, column=0, padx=10, pady=5, sticky="w")
roll_entry = tk.Entry(form_frame, font=("Helvetica", 12), width=30, bd=2)
roll_entry.grid(row=2, column=1, pady=5)

tk.Label(form_frame, text="Age:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=3, column=0, padx=10, pady=5, sticky="w")
age_entry = tk.Entry(form_frame, font=("Helvetica", 12), width=30, bd=2)
age_entry.grid(row=3, column=1, pady=5)

tk.Label(form_frame, text="Branch:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=4, column=0, padx=10, pady=5, sticky="w")
branch_combobox = ttk.Combobox(form_frame, font=("Helvetica", 12), width=28, state="readonly")
branch_combobox['values'] = ("EXTC", "IT", "CS", "Mechanical", "Civil")
branch_combobox.set("Select Branch")
branch_combobox.grid(row=4, column=1, pady=5)

# Buttons with styling
button_frame = tk.Frame(root, bg="#f0f0f0")
button_frame.pack(pady=10)

submit_button = tk.Button(button_frame, text="Submit", font=("Helvetica", 12, "bold"), bg="#4CAF50", fg="white", bd=2, relief="raised", width=12, command=submit_details)
submit_button.grid(row=0, column=0, padx=10)

clear_button = tk.Button(button_frame, text="Clear", font=("Helvetica", 12, "bold"), bg="#f44336", fg="white", bd=2, relief="raised", width=12, command=clear_fields)
clear_button.grid(row=0, column=1, padx=10)

# Button to download the receipt
# Button to download the receipt
download_button = tk.Button(root, text="Download Receipt", font=("Helvetica", 12, "bold"), bg="#008CBA", fg="white", bd=2, relief="raised", width=20, command=download_receipt)
download_button.pack(pady=10)  # Use pack instead of grid
download_button.pack_forget()  # Initially hide the download button


# Run the GUI event loop
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
import openpyxl
from openpyxl import Workbook
import os

# File to store the data
FILE_NAME = "Student_Details.xlsx"

# Global variable to store the generated PDF file path
pdf_filename = ""

# Function to generate a professional-looking PDF receipt with token number
def generate_pdf(name, roll_number, age, branch, token_number):
    global pdf_filename
    pdf_filename = f"Student_Receipt_{token_number}.pdf"
    
    c = canvas.Canvas(pdf_filename, pagesize=letter)
    
    # Add Header
    c.setFont("Helvetica-Bold", 18)
    c.drawString(180, 750, "Official Student Details Receipt")
    
    # Add Company/Institute Details (Logo or name can be added here)
    c.setFont("Helvetica", 12)
    c.drawString(100, 730, "Fr. C. Rodrigues Institute of Technology")
    c.drawString(100, 715, "Navi Mumbai, Maharashtra, India")

    # Add receipt details
    c.setFont("Helvetica-Bold", 14)
    c.drawString(100, 680, f"Name: {name}")
    c.drawString(100, 660, f"Roll Number: {roll_number}")
    c.drawString(100, 640, f"Age: {age}")
    c.drawString(100, 620, f"Branch: {branch}")
    c.drawString(100, 600, f"Token Number: {token_number}")
    
    # Add footer (receipt validity or disclaimer)
    c.setFont("Helvetica-Oblique", 10)
    c.drawString(100, 100, "This is an automatically generated receipt. Please keep it for future reference.")

    # Save the PDF
    c.save()

    messagebox.showinfo("PDF Generated", f"PDF receipt has been saved as {pdf_filename}.")
    return pdf_filename

# Function to save student details to Excel and assign token number
def save_to_excel(name, roll_number, age, branch):
    if not os.path.exists(FILE_NAME):
        # Create a new workbook and add a header row
        workbook = Workbook()
        sheet = workbook.active
        sheet.title = "Details"
        sheet.append(["Token Number", "Name", "Roll Number", "Age", "Branch"])
        workbook.save(FILE_NAME)

    # Open the workbook
    workbook = openpyxl.load_workbook(FILE_NAME)
    sheet = workbook.active

    # Assign a unique token number based on the number of rows
    token_number = sheet.max_row  # Token number is the current row number

    # Check for duplicate Roll Number
    for row in sheet.iter_rows(min_row=2, values_only=True):
        if row[2] == roll_number:
            return False, token_number

    # Append the data with token number
    sheet.append([token_number, name, roll_number, age, branch])  # Add data row with token number
    workbook.save(FILE_NAME)
    return True, token_number

# Function to collect and display student details
def submit_details():
    name = name_entry.get()
    roll_number = roll_entry.get()
    age = age_entry.get()
    branch = branch_combobox.get()

    if not (name and roll_number and age and branch) or branch == "Select Branch":
        messagebox.showwarning("Input Error", "All fields are required!")
        return

    # Save details to Excel and check for duplicates
    is_saved, token_number = save_to_excel(name, roll_number, age, branch)
    if is_saved:
        messagebox.showinfo("Success", "Your contact has been updated!")
        
        # Generate PDF receipt with token number
        generate_pdf(name, roll_number, age, branch, token_number)

        clear_fields()
        ask_download(token_number)  # Ask if the user wants to download the PDF
    else:
        messagebox.showwarning("Duplicate Entry", "This record already exists!")

# Function to clear input fields
def clear_fields():
    name_entry.delete(0, tk.END)
    roll_entry.delete(0, tk.END)
    age_entry.delete(0, tk.END)
    branch_combobox.set("Select Branch")

# Function to ask if the user wants to download the PDF
def ask_download(token_number):
    response = messagebox.askquestion("Download Receipt", f"Do you want to download the receipt for token {token_number}?")
    if response == "yes":
        download_receipt()

# Function to download the PDF receipt
def download_receipt():
    if pdf_filename:
        messagebox.showinfo("Download Receipt", f"Download your receipt from: {os.path.abspath(pdf_filename)}")

# Function to handle pressing Enter to move to the next field
def focus_next_field(event):
    current_widget = event.widget
    widget_list = [name_entry, roll_entry, age_entry, branch_combobox]
    next_widget = widget_list[widget_list.index(current_widget) + 1] if widget_list.index(current_widget) + 1 < len(widget_list) else submit_button
    next_widget.focus_set()

# Create main window
root = tk.Tk()
root.title("Student Details Form")
root.geometry("400x500")
root.configure(bg="#f0f0f0")  # Set background color

# Create a frame for the form elements
form_frame = tk.Frame(root, bg="#f0f0f0")
form_frame.pack(pady=20)

# Add header label
header_label = tk.Label(form_frame, text="Student Details Form", font=("Helvetica", 16, "bold"), bg="#f0f0f0")
header_label.grid(row=0, column=0, columnspan=2, pady=10)

# Create labels and entry fields with styling
tk.Label(form_frame, text="Name:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=1, column=0, padx=10, pady=5, sticky="w")
name_entry = tk.Entry(form_frame, font=("Helvetica", 12), width=30, bd=2)
name_entry.grid(row=1, column=1, pady=5)
name_entry.bind("<Return>", focus_next_field)  # Bind Enter key to move to next field

tk.Label(form_frame, text="Roll Number:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=2, column=0, padx=10, pady=5, sticky="w")
roll_entry = tk.Entry(form_frame, font=("Helvetica", 12), width=30, bd=2)
roll_entry.grid(row=2, column=1, pady=5)
roll_entry.bind("<Return>", focus_next_field)

tk.Label(form_frame, text="Age:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=3, column=0, padx=10, pady=5, sticky="w")
age_entry = tk.Entry(form_frame, font=("Helvetica", 12), width=30, bd=2)
age_entry.grid(row=3, column=1, pady=5)
age_entry.bind("<Return>", focus_next_field)

tk.Label(form_frame, text="Branch:", font=("Helvetica", 12), bg="#f0f0f0").grid(row=4, column=0, padx=10, pady=5, sticky="w")
branch_combobox = ttk.Combobox(form_frame, font=("Helvetica", 12), width=28, state="readonly")
branch_combobox['values'] = ("EXTC", "IT", "CS", "Mechanical", "Civil")
branch_combobox.set("Select Branch")
branch_combobox.grid(row=4, column=1, pady=5)
branch_combobox.bind("<Return>", focus_next_field)

# Buttons with styling
button_frame = tk.Frame(root, bg="#f0f0f0")
button_frame.pack(pady=10)

submit_button = tk.Button(button_frame, text="Submit", font=("Helvetica", 12, "bold"), bg="#4CAF50", fg="white", bd=2, relief="raised", width=12, command=submit_details)
submit_button.grid(row=0, column=0, padx=10)

clear_button = tk.Button(button_frame, text="Clear", font=("Helvetica", 12, "bold"), bg="#f44336", fg="white", bd=2, relief="raised", width=12, command=clear_fields)
clear_button.grid(row=0, column=1, padx=10)

# Run the GUI event loop
root.mainloop()
```
